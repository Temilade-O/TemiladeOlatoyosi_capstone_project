```python
age_ranges = ["[{0} - {1})".format(age, age + 10) for age in range(20, 60, 10)]
age_ranges
count_unique_age_ranges = len(age_ranges)
count_unique_age_ranges
df['age_range'] = pd.cut(x=df['Age'], bins=count_unique_age_ranges, labels=age_ranges)
df.head()
```

```python
empatrr = df.drop(['DistanceFromHome', 'EmployeeCount', 'EmployeeNumber', 'HourlyRate', 'Over18', 'StandardHours', 'StockOptionLevel', 'YearsInCurrentRole','DailyRate', 'YearsSinceLastPromotion', 'YearsWithCurrManager'], axis=1)
empatrr.head()
```

```python
df_empattr= pd.DataFrame(empatrr)
df_empattr['age_range'].value_counts()
```

```python
#the relationship between attrition and age-range and education level
sns.catplot(x='age_range', y='Education', data= df_empattr, hue='Attrition', kind= 'bar' )
plt.show()
```

```python
sns.pairplot(df_empattr, hue='Attrition', height=2.5);
```
